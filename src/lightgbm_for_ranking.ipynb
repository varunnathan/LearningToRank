```python
# check if queries are in sorted order

out = []
queries = []
cnt = 0
line_cnt = 0
last_qid = -1
for line in open(TRAIN_FN, 'r'):
    if line_cnt % 100 == 0:
        print('# lines read: %d' % (line_cnt))
    tokens = line.strip().split(' ')
    qid = tokens[1].split(':')[1]
    queries.append(qid)
    
    if line_cnt == 0:
        # first line
        last_qid = qid
    
    if qid != last_qid:
        out.append({'qid': last_qid, 'cnt': cnt})
        last_qid = qid
        cnt = 0
    
    cnt += 1
    line_cnt += 1

print('Total # lines: %d' % (line_cnt))
out.append({'qid': qid, 'cnt': cnt})
out_df = pd.DataFrame(out)
```

```python
# convert txt files into libsvm format

def convert(input_filename, out_data_filename, out_query_filename,
            out_query_filename2):
    inp = open(input_filename, 'r')
    out_feature = open(out_data_filename, 'w')
    out_query = open(out_query_filename, 'w')
    out_query2 = open(out_query_filename2, 'w')
    
    cur_cnt, cur_doc_cnt = 0, 0
    last_qid = -1
    
    while True:
        line = inp.readline()
        
        if not line:
            break
            
        tokens = line.strip().split(' ')
        label = tokens[0]
        qid = tokens[1].split(':')[1]
        
        if last_qid != qid:
            if cur_doc_cnt > 0:
                out_query.write(str(cur_doc_cnt) + '\n')
                out_query2.write(str(cur_doc_cnt) + '\n')
                cur_cnt += 1
            cur_doc_cnt = 0
            last_qid = qid
            
        cur_doc_cnt += 1
        out_feature.write(label + ' ')
        out_feature.write(' '.join(tokens[2:]) + '\n')
        
    # last query
    out_query.write(str(cur_doc_cnt) + '\n')
    out_query2.write(str(cur_doc_cnt) + '\n')
    
    inp.close()
    out_feature.close()
    out_query.close()
    out_query2.close()
```

### lightgbm on pandas dataframe

```python
def convert_txt_to_df(txt_fn):
    inp = open(txt_fn, 'r')
    
    labels, feats = [], []
    for i, line in enumerate(inp):
        if i % 100000 == 0:
            print('# lines read: %d' % (i))
        tokens = line.strip().split(' ')
        label = tokens[0]
        feat = {}
        for token in tokens[1:]:
            d = {int(token.split(':')[0]): float(token.split(':')[1])}
            feat.update(d)
        labels.append(label)
        feats.append(feat)
    
    return labels, feats
```
